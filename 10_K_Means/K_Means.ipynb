## K-Means

O algoritmo K-means é um método de clustering que agrupa dados em k grupos. Inicia-se escolhendo k pontos iniciais como centroides, representantes do centro de cada cluster. Em seguida, atribui cada ponto de dados ao centroide mais próximo, formando clusters iniciais. Os centroides são recalculados como a média dos pontos pertencentes a cada cluster. Esse processo de atribuição e atualização é repetido até que os centroides não mudem significativamente. O objetivo é minimizar a variância intra-cluster, resultando em clusters coesos. 

O K-means possui limitações, como sensibilidade à escolha inicial dos centroides e a necessidade de especificar k antecipadamente, para suprir esse problema temos o metodo **K-Meanss ++**.

## Maneiras de definir o K

### 1. K-Means ++

O K-means++ é uma melhoria do algoritmo K-means que aprimora a escolha inicial dos centroides. Começa escolhendo o primeiro centróide aleatoriamente e, em seguida, seleciona os próximos centroides com base na distância aos centroides já escolhidos, no caso a probabilidade de seleção do próximo centróide é proporcional ao quadrado da distância ao centróide mais próximo já escolhido.

Esse método distribuir melhor os centroides inicialmente, tornando o algoritmo mais robusto e melhorando a qualidade dos clusters resultantes. Essa abordagem é eficaz na redução da sensibilidade à escolha inicial, contribuindo para convergência mais rápida e resultados mais estáveis.

### 2. Regra prática

É considerado uma regra informal, usada para tomar decisões rápidas ou fazer estimativas aproximadas, sem a necessidade de cálculos detalhados ou métodos formais.

$$ \text{Número de Clusters} = \sqrt{\frac{n}{2}}$$

$n$: Número de observações do dataset.

### 3. Método de Elbow

É uma técnica usada para determinar o número ideal de clusters em algoritmos de clustering. Ele envolve executar o K-means para diferentes valores de k, calcular a variação da Soma dos Quadrados Intra-Cluster e identificar o "cotovelo" no gráfico, indicando o número ótimo de clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Gerar dados fictícios
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Calcular WCSS para diferentes valores de k
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_) # Soma dos Quadrados Intra-Cluster

# Plotar gráfico Elbow
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Método de Elbow para Escolha de K')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Soma dos Quadrados Intra-Cluster')
plt.show()
```

O "cotovelo" é considerado um indicativo de que o aumento no número de clusters não contribui significativamente para a redução da variância intra-cluster, no exemplo acimo 4 cluster já é o ideal.
